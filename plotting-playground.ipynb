```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Generate random true and predicted arrays
true_labels = np.random.choice([0, 1, 2], size=100)
pred_labels = np.random.choice([0, 1, 2], size=100)

# Compute confusion matrix
cm = confusion_matrix(true_labels, pred_labels)

# Plot confusion matrix
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', cbar=False, ax=ax)

# Customize the plot
ax.set_title('Confusion Matrix', fontsize=16, pad=20)
ax.set_xlabel('Predicted Labels', fontsize=14)
ax.set_ylabel('True Labels', fontsize=14)
ax.set_xticklabels(['Class 0', 'Class 1', 'Class 2'], fontsize=12)
ax.set_yticklabels(['Class 0', 'Class 1', 'Class 2'], fontsize=12, rotation=0)

plt.show()
```

```python
import matplotlib.pyplot as plt

# GPA values for the semesters
semesters = list(range(1, 9))
gpa = [3.64, 3.64, 3.74, 3.91, 3.96, 4.00, None, None]

# Plot the GPA values
plt.figure(figsize=(10, 6))
plt.plot(semesters, gpa, marker='o', linestyle='-', color='b', label='GPA per Semester')

# Highlight the CGPA
plt.axhline(y=cgpa, color='black', linestyle='--', label=f'CGPA: {cgpa:.2f}')

# Annotate the points with their values
for i, g in enumerate(gpa):
    if g is not None:
        plt.text(semesters[i], g, f'{g}', fontsize=12, verticalalignment='bottom', horizontalalignment='right')

# Set the labels and title
plt.xlabel('Semester')
plt.ylabel('GPA')
plt.title('GPA per Semester and CGPA')
plt.legend()

# Customize x-axis and y-axis
plt.xticks(semesters)
plt.ylim(3.5, 4.05)

# Show the plot
# plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# User's GPA values for the semesters
semesters = list(range(1, 9))
gpa_user = [3.64, 3.64, 3.74, 3.91, 3.96, 4.0, None, None]
cgpa_user = np.mean([x for x in gpa_user if x is not None])

# Friend's GPA values for the semesters
gpa_friend = [3.85, 3.76, 3.84, 3.76, 3.93, 4.0, None, None]
cgpa_friend = np.mean([x for x in gpa_friend if x is not None])


# Replace None with zeros for area plot
gpa_user_filled = [g if g is not None else 0 for g in gpa_user]
gpa_friend_filled = [g if g is not None else 0 for g in gpa_friend]

# Create an area plot
plt.figure(figsize=(10, 6))
plt.fill_between(semesters, gpa_user_filled, color='skyblue', alpha=0.5, label=f'User {cgpa_user:.2f}')
plt.plot(semesters, gpa_user_filled, color='blue', marker='o')

plt.fill_between(semesters, gpa_friend_filled, color='lightgreen', alpha=0.5, label=f"Friend {cgpa_friend:.2f}")
plt.plot(semesters, gpa_friend_filled, color='green', marker='o')

# Highlight CGPA
# plt.axhline(y=cgpa_user, color='red', linestyle='--', label='User CGPA: 3.78')
# plt.axhline(y=cgpa_friend, color='orange', linestyle='--', label="Friend's CGPA: 3.83")

# Annotate points with GPA values
for i, g in enumerate(gpa_user):
    if g is not None:
        plt.text(semesters[i], g, f'{g}', fontsize=12, verticalalignment='bottom', horizontalalignment='right')

for i, g in enumerate(gpa_friend):
    if g is not None:
        plt.text(semesters[i], g, f'{g}', fontsize=12, verticalalignment='bottom', horizontalalignment='left')

# Set the labels and title
plt.xlabel('Semester')
plt.ylabel('GPA')
plt.title('GPA per Semester and CGPA for User and Friend')
plt.legend(loc='best')

# Customize x-axis and y-axis
plt.xticks(semesters)
plt.ylim(3.5, 4.05)

# Show the plot
# plt.grid(True)
plt.tight_layout()
plt.show()
```
